## Model

```python
frequent_itemsets = apriori(df,min_support =0.05,use_colnames=True)
frequent_itemsets.sort_values('support',ascending=False)
```

### association rules

```python
# metrices = ['support', 'confidence', 'lift', 'leverage', and 'conviction']

metric='lift'
# rules = association_rules(frequent_itemsets, metric = "support", min_threshold = .2)
# rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = .5)
rules = association_rules(frequent_itemsets, metric = metric, min_threshold = 1)

rules.sort_values('lift', ascending = False, inplace = True)
rules
```

การที่เรียงข้อมูลตามค่า lift มาก แสดงว่ากลุ่มสินค้า antecedent นั้นมีความเกี่ยวข้องหรือมีความสัมพันธ์กับกลุ่มสินค้า consequent สูง  
เมื่อลูกค้าหยิบกลุ่มสินค้า antecedent ลงตะกร้าไปแล้ว มีโอกาสที่ลูกค้าจะหยิบกลุ่มสินค้า consequent เพิ่มมากขึ้นตามค่า lift นั้นๆ  
* ตัวอย่างเช่น  
    - เมื่อลูกค้าหยิบ Dill,Cheese โอกาสที่จะหยิบ Kidney Beans,Onion มีค่า 1.826002 เท่า  
    - เมื่อลูกค้าหยิบ Dill โอกาสที่จะหยิบ Bread,Butter มีค่า 1.004020 เท่า  

ดังนั้นหากต้องการให้ยอดขายเพิ่มขึ้นอีก ควรที่จะจัดสินค้าที่มีความสัมพันธ์สูงเอาไว้ด้วยกัน เพื่อที่จะให้ลูกค้ามองเห็นได้ง่ายและมีโอกาสที่ลูกค้าจะหยิบสินค้าไปด้วยกันมากขึ้น  หรือจะเป็นการจัดโปรโมชั่นระหว่างสินค้าที่มีความสัมพันธ์กันสูง เพื่อที่ง่ายต่อการตัดสินใจของลูกค้า
